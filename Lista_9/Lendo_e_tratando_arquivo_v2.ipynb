```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Supondo que seu DataFrame seja 'df'

# 1. Calcular a matriz de correlação usando SPEARMAN (Ideal para seus outliers)
matriz_correlacao = baseOriginal.corr(method='spearman')

# 2. Visualizar com um Heatmap (Mapa de Calor)
plt.figure(figsize=(20, 20))
sns.heatmap(matriz_correlacao, annot=True, fmt=".2f", cmap='coolwarm', vmin=-1, vmax=1)
plt.title("Mapa de Correlação de Spearman (Robusto a Outliers)")
plt.show()
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
#Verificando a multicolinearidade

# O VIF não lida bem com valores infinitos ou NaN, então limpe antes
X_vif = baseOriginal.drop(columns=['Class']) # Remova a variável alvo (fraude)
X_vif = X_vif.select_dtypes(include=['float64', 'int64']) # Apenas numéricos

# Função para calcular VIF
vif_data = pd.DataFrame()
vif_data["feature"] = X_vif.columns
vif_data["VIF"] = [variance_inflation_factor(X_vif.values, i)
                          for i in range(len(X_vif.columns))]

# Ordenar para ver os piores casos primeiro
print(vif_data.sort_values(by="VIF", ascending=False))
```

```python
baseOriginal = baseOriginal.drop(columns='Amount')
```

```python
Nome_das_Colunas = baseOriginal.columns[:-2]
print(Nome_das_Colunas)
```

```python
df_duplicates = baseOriginal[baseOriginal.duplicated(subset=Nome_das_Colunas,keep=False)]
# se houver valores redundantes ou inconsistentes, imprima
if len(df_duplicates)>0:
    print('\nAmostras redundantes ou inconsistentes:')
    print(len(df_duplicates))
    display(df_duplicates)
else:
    print('Não existem valores duplicados')
```

```python
def delDuplicatas( df_dataset ):
    '''
    Para cada grupo de amostras duplicadas, mantém uma e apaga as demais
    '''

    # remove as amostras duplicadas, mantendo apenas a primeira ocorrencia
    df_dataset = df_dataset.drop_duplicates(keep = 'first')

    return df_dataset

basePP_Red = delDuplicatas( baseOriginal )
print(len(baseOriginal))
print(len(basePP_Red))
```

```python
# para detectar inconsistências, a rotina abaixo obtém as amostras onde os valores
# dos atributos continuam duplicados. Neste caso, os atributos serão iguais, mas as classes serão distintas
df_duplicates = basePP_Red[basePP_Red.duplicated(subset=Nome_das_Colunas,keep=False)]

# se tiver valores inconsistentes, imprime
if len(df_duplicates)>0:
    print('\nAmostras inconsistentes:')
    print(len(df_duplicates))
    display(df_duplicates)
else:
    print('Não existem mostras inconsistentes')
```

```python
from sklearn.model_selection import train_test_split
X_prev = basePP_Red.drop(columns='Class')
Y_prev = basePP_Red['Class']


X_treino, X_teste, y_treino, y_teste = train_test_split(X_prev, Y_prev, test_size = 0.20, random_state = 42)
```

```python
zscore = StandardScaler()

y_treino = y_treino.to_frame()
y_teste = y_teste.to_frame()


X_treino_outl = pd.DataFrame(zscore.fit_transform(X_treino), columns=X_treino.columns)
X_teste_outl = pd.DataFrame(zscore.fit_transform(X_teste), columns=X_teste.columns)
X_treino
```

```python
from imblearn.under_sampling import TomekLinks
from imblearn.over_sampling import SMOTE

smote = SMOTE(random_state=42)
tlink = TomekLinks(sampling_strategy='auto')
y_treino = y_treino.astype(int)
y_teste = y_teste.astype(int)

X_treino_outl_bal, y_treino_bal = smote.fit_resample(X_treino_outl, y_treino)
#X_treino_outl_bal, y_treino_bal = tlink.fit_resample(X_treino_outl_bal, y_treino_bal)
```
