```python
from minisom import MiniSom
import numpy as np
from sklearn.metrics import silhouette_score

grid_sizes = [(1,2), (2,1)]

X_numpy = X_treino.values

for h, w in grid_sizes:
    print(h)
    print(w)
    
    som = MiniSom(
        x=h, 
        y=w,
        input_len=X_numpy.shape[1],
        sigma=1.0,
        learning_rate=0.5,
        neighborhood_function='gaussian',
        random_seed=42
    )
    som.train_random(X_numpy, len(X_numpy))
    bmus = np.array([som.winner(x) for x in X_numpy])
    labels = np.array([i * w + j for (i, j) in bmus])
    num_clusters = len(np.unique(labels))

    if num_clusters == 2:
        try:
            sil = silhouette_score(X_numpy, labels, sample_size=50000)
            print(f"SOM {h}x{w}: 2 clusters, silhouette={sil:.4f}")
        except:
            print(f"SOM {h}x{w}: 2 clusters, mas não foi possível calcular silhouette")

```
