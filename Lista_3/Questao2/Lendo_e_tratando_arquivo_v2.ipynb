```python
col_drop = ['PassengerId', 'Name', 'Ticket', 'Cabin']
base.drop(columns= col_drop, inplace=True)
base
```



>** OneHotEncoder - Agora vamos binarizar atributos não ordinais**

```python
col_median_null_values = ['Age']
col_mode_null_values = ['Embarked']
base[col_median_null_values] = base[col_median_null_values].fillna(base[col_median_null_values].median())
base[col_mode_null_values] = base[col_mode_null_values].fillna(base[col_mode_null_values].mode().iloc[0])
base
```

```python
cols_labels_encode = ['Sex']
cols_ordinal_encode = ['Embarked']
values_embarked = ['S', 'C', 'Q']
base[cols_labels_encode] = base[cols_labels_encode].apply(LabelEncoder().fit_transform)
embarked_encoder = OrdinalEncoder(categories=[values_embarked])
base[cols_ordinal_encode] = embarked_encoder.fit_transform(base[cols_ordinal_encode]).astype(int)
```
